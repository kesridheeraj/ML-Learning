```python
import pandas as pd
import numpy as np
from IPython.display import display
from sklearn.model_selection import train_test_split
auto_data = pd.read_csv('Data/02/demos/data/imports-85.data', sep=',', engine='python')
auto_data.head(10)
```

```python
auto_data = auto_data.replace('?', np.nan)
auto_data.head(5)
```

```python
auto_data.columns[auto_data.dtypes.eq('object')]
```

```python
auto_data = auto_data.drop('normalized-losses', axis = 1)
auto_data.head(10)
```

```python
columns_to_make_numeric = ['price', 'bore', 'stroke', 'horsepower', 'peak-rpm']
auto_data[columns_to_make_numeric] = auto_data[columns_to_make_numeric].apply(pd.to_numeric, errors = 'coerce')
auto_data.columns[auto_data.dtypes.eq('object')]
```

```python
categorical_columns = ['make', 'fuel-type', 'aspiration', 'num-of-doors', 'body-style', 
                       'drive-wheels', 'engine-location', 'engine-type', 'fuel-system']
auto_data = pd.get_dummies(auto_data, columns = categorical_columns)
pd.options.display.max_columns = None
auto_data.head(10)
```

```python
num_of_cylinders_mapping = {'two': 2, 'three': 3, 'four': 4, 'five': 5, 'six': 6, 'eight': 8, 'twelve': 12}
auto_data['num-of-cylinders'].replace(num_of_cylinders_mapping, inplace = True)
auto_data.head(10)
```

```python
auto_data = auto_data.dropna()
len(auto_data)
```

```python
auto_data[auto_data.isnull().any(axis = 1)]
```

```python
X = auto_data.drop('price', axis = 1)
Y = auto_data['price']
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2)
```

```python
from sklearn.linear_model import LinearRegression
linear_regression_model = LinearRegression()
linear_regression_model.fit(X_train, Y_train)
```

```python
linear_regression_model.score(X_train, Y_train)
```

```python
predictors = X_train.columns
predictors_weight = pd.Series(linear_regression_model.coef_, predictors)
predictors_weight.sort_values()
```

```python
linear_regression_model.predict(X_test)
linear_regression_model.score(X_test, Y_test)
```
